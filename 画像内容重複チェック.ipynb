```python
import os
import cv2
import shutil

old_f = "old_pic"
new_f = "new_pic"
ok_f = "check_ok"
ng_f = "check_ng"

oldFiles = os.listdir(old_f)
oldSizeList = []

for f in oldFiles:
    img = cv2.imread(old_f + os.sep + f)
    if img is None:
        print(f + " 読込エラー")
        continue
    
    oldSizeList.append([img.shape[0], img.shape[1]])

print("oldSizeList完了")
```

```python
nowCount = 0
for file in os.listdir(new_f):
    target = new_f + os.sep + file
    
    img = cv2.imread(target)
    if img is None:
        print(file + " 読込エラー")
        continue
    height = img.shape[0]
    width = img.shape[1]
    i = 0
    
    isSameAri = False
    sameImgName = ""
    for size in oldSizeList:
        if size[0] == height and size[1] == width:
            checkImg = cv2.imread(old_f + os.sep + oldFiles[i])
            isFuicchi = False
            for row in range(height):
                for col in range(width):
                    if not (img[row, col] == checkImg[row, col]).all():
                        isFuicchi = True
                        break
                if isFuicchi:
                    break
            
            if not isFuicchi:
                isSameAri = True
                sameImgName = oldFiles[i]
                break
            
        i += 1
        
    if isSameAri:
        shutil.move(target, ng_f + os.sep + file)
        print(file + " = " + sameImgName)
    else:
        shutil.move(target, ok_f + os.sep + file)
    
    nowCount += 1
    if nowCount % 100 == 0:
        print(str(nowCount) + "件完了")
```
